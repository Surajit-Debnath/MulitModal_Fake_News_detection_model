```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("fakeddit_cleaned.csv")
```

```python
print(df.shape)

df.head()
```

```python
df["2_way_label"].value_counts()
```

```python
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    random_state=42,
    stratify=df["2_way_label"]
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    random_state=42,
    stratify=temp_df["2_way_label"]
)
```

```python
print("Training samples   :", len(train_df))
print("Validation samples :", len(val_df))
print("Testing samples    :", len(test_df))
```

```python
print("Train")
print(train_df["2_way_label"].value_counts(normalize=True))

print("\nValidation")
print(val_df["2_way_label"].value_counts(normalize=True))

print("\nTest")
print(test_df["2_way_label"].value_counts(normalize=True))
```

```python
train_df.to_csv("data/processed/train.csv", index=False)
val_df.to_csv("data/processed/validation.csv", index=False)
test_df.to_csv("data/processed/test.csv", index=False)
```
